```
import numpy as np
import matplotlib.pyplot as plt

# Data
square_footage = np.array([1200, 1400, 1600, 1800, 2000, 2200, 2400, 2600, 2800, 3000])
selling_prices = np.array([250, 290, 315, 380, 410, 450, 500, 525, 570, 610])

# Create scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(square_footage, selling_prices, color='green')
plt.title('Housing Price vs. Square Footage')
plt.xlabel('Square Footage (sq.ft.)')
plt.ylabel('Selling Price ($000s)')
plt.grid(True)
plt.show()
```

Lab4: Suppose you're a sales manager for an e-commerce company, and you

electronics_sales = np.array([25000,28000,31000,27000,30000, 32000, 35000,36000, 38000, 39000, 41000, 42000])

```
import numpy as np
import matplotlib.pyplot as plt

# Data
months = np.arange(1, 13)
electronics_sales = np.array([25000, 28000, 31000, 27000, 30000, 32000, 35000, 36000, 38000, 39000, 41000, 42000])
clothing_sales = np.array([15000, 16000, 17000, 18000, 19000, 20000, 21000, 22000, 23000, 24000, 25000, 26000])
home_garden_sales = np.array([18000, 19000, 20000, 21000, 22000, 23000, 24000, 25000, 26000, 27000, 28000, 29000])
sports_outdoors_sales = np.array([12000, 13000, 14000, 15000, 16000, 17000, 18000, 19000, 20000, 21000, 22000, 23000])

# Create subplots
fig, axs = plt.subplots(2, 2, figsize=(15, 10))

# Plot for Electronics
axs[0, 0].plot(months, electronics_sales, marker='o', color='b')
axs[0, 0].set_title('Electronics')
axs[0, 0].set_xlabel('Month')
axs[0, 0].set_ylabel('Sales Amount')

# Plot for Clothing
axs[0, 1].plot(months, clothing_sales, marker='o', color='g')
axs[0, 1].set_title('Clothing')
axs[0, 1].set_xlabel('Month')
axs[0, 1].set_ylabel('Sales Amount')

# Plot for Home & Garden
axs[1, 0].plot(months, home_garden_sales, marker='o', color='r')
axs[1, 0].set_title('Home & Garden')
axs[1, 0].set_xlabel('Month')
axs[1, 0].set_ylabel('Sales Amount')

# Plot for Sports & Outdoors
axs[1, 1].plot(months, sports_outdoors_sales, marker='o', color='m')
axs[1, 1].set_title('Sports & Outdoors Sales')
axs[1, 1].set_xlabel('Month')
axs[1, 1].set_ylabel('Sales Amount')

# Set common labels
fig.suptitle('Sales Performance of Different Product Categories Over Time', fontsize=20)

# Show plot
plt.show()
```

Scenario:Analyzing Sales Data

Suppose you work for a retail company, and you have dummy data containing sales
data for the past year. The data includes information such as SalesDate,product
names,regions, sales quantities, prices, and dates. You have to generate a bar chart,pie plot on region and prices and line chart on SalesDate and prices columns.

Further, you need to get some inference out of the chart.

Create a ChatGPT prompt to generate the code for this scenario. Based on the code generated, ask ChatGPT to give the conclusion/inference.

Note: You can provide the data to ChatGPT or ask it to use sample data.

```
import numpy as np
import matplotlib.pyplot as plt

# Sample data
sales_dates = np.arange('2023-01', '2024-01', dtype='datetime64[M]')
products = np.array(['Product A', 'Product B', 'Product C', 'Product A', 'Product B', 'Product C',
                     'Product A', 'Product B', 'Product C', 'Product A', 'Product B', 'Product C'])
regions = np.array(['North', 'South', 'East', 'West', 'North', 'South', 'East', 'West', 'North',
                    'South', 'East', 'West'])
quantities = np.array([120, 150, 100, 130, 160, 110, 140, 170, 120, 130, 150, 100])
prices = np.array([200, 300, 150, 220, 320, 160, 240, 340, 180, 250, 350, 170])

# Bar chart for total sales quantity for each product
unique_products = np.unique(products)
product_sales = np.array([np.sum(quantities[products == product]) for product in unique_products])
plt.figure(figsize=(10, 6))
plt.bar(unique_products, product_sales, color='skyblue')
plt.title('Total Sales Quantity for Each Product')
plt.xlabel('Product')
plt.ylabel('Total Quantity Sold')
plt.show()

# Pie chart for distribution of total sales prices by region
unique_regions = np.unique(regions)
region_prices = np.array([np.sum(prices[regions == region]) for region in unique_regions])
plt.figure(figsize=(8, 8))
plt.pie(region_prices, labels=unique_regions, autopct='%1.1f%%', colors=['#ff9999','#66b3ff','#99ff99','#ffcc99'])
plt.title('Distribution of Total Sales Prices by Region')
plt.show()

# Line chart for sales prices over time
plt.figure(figsize=(12, 6))
plt.plot(sales_dates, prices, marker='o', color='b')
plt.title('Sales Prices Over Time')
plt.xlabel('Sales Date')
plt.ylabel('Sales Price')
plt.grid(True)
plt.show()
```

#Analysis/Conclusion:
1. Bar Chart Analysis:

* This chart shows the total sales quantity for each product. By comparing the heights of the bars, you can identify which product had the highest and lowest sales quantity.
2. Pie Chart Analysis:

* This pie chart represents the distribution of total sales prices by region. The chart provides a visual representation of the contribution of each region to the total sales prices, allowing you to see which region had the highest sales prices.
3. Line Chart Analysis:

* The line chart illustrates the trend of sales prices over the past year. It helps identify patterns or trends in the sales prices over time, such as seasonality or significant changes in prices.
